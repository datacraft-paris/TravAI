### "-" handling UPDATED

Since a lot of values are missing, we won't be handling them by hand. Values could be filled with an LLM but it requires a significative amount of API calls, which is against frugal AI practices (we need a big model with general knowledge to make good estimations).

The prefered approach is to correct the missing data furthur in the pipeline : when an ingredient is retrieved, if a value is missing, an API call will be made to infere the value.

### "<0,5 handling"

ChatGPT recommends an approach where we divide by two the number. It will keep overall mean and variance closer to reality, and this practice is often used in food and health.

```python
def replace_less_than_values(value):
    # Check if the value matches the pattern "< X"
    value = str(value)
    match = re.match(r'< (\d+,\d+|\d+)', value)
    if match:
        # Extract the number and convert it to a float
        number_str = match.group(1).replace(',', '.')  # Handle comma as decimal separator
        number = float(number_str)
        number_divided_by_two = number / 2
        # Return the number divided by two
        return str(number_divided_by_two)
    return value
```

```python
list_of_columns_with_less_than_values_to_replace = ['Energie, N x facteur Jones, avec fibres  (kcal/100 g)','Eau (g/100 g)', 'Protéines, N x facteur de Jones (g/100 g)',
       'Protéines, N x 6.25 (g/100 g)', 'Glucides (g/100 g)',
       'Lipides (g/100 g)', 'Sucres (g/100 g)','Cholestérol (mg/100 g)', 'Sel chlorure de sodium (g/100 g)',
       'Calcium (mg/100 g)', 'Fer (mg/100 g)', 'Magnésium (mg/100 g)',
       'Vitamine D (µg/100 g)',
       'Vitamine C (mg/100 g)', 'Vitamine B9 ou Folates totaux (µg/100 g)',
       'Vitamine B12 (µg/100 g)']
```

```python
for field in list_of_columns_with_less_than_values_to_replace :
  df_filtered[field] = df_filtered[field].apply(replace_less_than_values)
  print(df_filtered[field].value_counts())
```

```python
df_filtered = df_filtered.dropna(subset=list_of_columns_with_less_than_values_to_replace)
```

```python
def translate_food_name(food_name):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",  # Use GPT-4o mini
            messages=[
                {"role": "system", "content": "You are a professional food translator. Translate French food names to accurate English equivalents. Just purely translate, DO NOT add extra text./n exemple : 'Oie, viande et peau, rôtie/cuite au four' becomes 'Goose, meat and skin, roasted/baked'"},
                {"role": "user", "content": f"Translate this food name from French to English: {food_name}"}
            ],
            temperature=0  # Ensures a deterministic response
        )
        return response["choices"][0]["message"]["content"].strip()

    except Exception as e:
        print(f"Error translating '{food_name}': {e}")
        return None  # Return None if the API call fails
```

```python
# Apply the translation function to each row, with a delay to avoid rate limits
df_filtered["alim_nom_en"] = df_filtered["alim_nom_fr"].apply(lambda x: translate_food_name(x) if isinstance(x, str) else None)
```
